```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df = pd.read_csv("Customer Churn.csv")
df.head()
```

```python
df.info()
```

##### Here we see that TotalCharges is in object Datatype , which should be FLOAT 
##### And the TotalCharge column has Blank values, changing them for 0

```python
df['TotalCharges']=df['TotalCharges'].replace(" ",0)
df['TotalCharges']=df['TotalCharges'].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

##### So as we can see out data has no null values

```python
df.describe()
```

Making sure there are no Duplicates in our data by checking duplicates for our primary key

```python
df["customerID"].duplicated().sum()
```

Converting values of senior citizen to Yes and No from 1 and 0 respectively 

```python
def convert(value):
    if value == 1 :
        return "Yes"
    else :
        return "No"
df["SeniorCitizen"]=df["SeniorCitizen"].apply(convert)
```

```python
df.head()
```

```python
ax = sns.countplot(x = df["Churn"])
ax.bar_label(ax.containers[0])
plt.title("Count of Customers by Churn")
plt.show()
```

```python
plt.figure(figsize = (4,4))
gb = df.groupby("Churn").agg({'Churn':"count"})
plt.title("Percentage of Churn Customers" , fontsize= 10 )
plt.pie(gb['Churn'], labels = gb.index , autopct = "%1.2f%%")
plt.show()
```

From the given piue chart we can conclude that 26.54% of our customers have turned out, now we will explore the reason 
behind it                                                       
Now lets explore things by gender

```python
plt.figure(figsize = (4,4))
sns.countplot(x= "gender",data = df , hue = "Churn")
plt.title("Churn by Gender")
plt.show()
```

```python

plt.figure(figsize = (4,4))
gb = df.groupby("SeniorCitizen").agg({'SeniorCitizen':"count"})
plt.title("Percentage Senior Citizens in total " , fontsize= 10 )
plt.pie(gb['SeniorCitizen'], labels = gb.index , autopct = "%1.2f%%")
plt.show()
```

```python
total_counts = df.groupby('SeniorCitizen')['Churn'].value_counts(normalize=True).unstack() * 100

#plot
fig , ax = plt.subplots(figsize=(4,4))
total_counts.plot(kind='bar' , stacked = True , ax=ax, color = ['#1f77b4','#ff7f0e']) #customizing color

for p in ax.patches:
    width,height = p.get_width(),p.get_height()
    x,y=p.get_xy()
    ax.text(x + width / 2 , y + height / 2 , f'{height:.1f}%',ha='center',va='center')

plt.title('Churn by SeniorCitizen')
plt.xlabel('SeniorCitizen')
plt.ylabel('Percentage')
plt.xticks(rotation=0)
plt.legend(title='Churn',bbox_to_anchor=(0.9,0.9)) #legend location

plt.show()
```

Comperatively a siginificantly greater number of senior citizens have churned out.

```python
plt.figure(figsize=(9,4))
sns.histplot(x='tenure',data=df,bins = 72,hue ='Churn')
plt.show()
```

The most churned customers are in the early stages 

```python
plt.figure(figsize=(4,4))
sns.countplot(x = 'Contract',data=df,hue='Churn')
plt.show()
```

So as we can tell customers on one month contract bases are most likely to churn compared to those who have yearly contracts.

```python
df.columns
```

```python
#Creating Subplots for our different services 
columns = ['PhoneService', 'MultipleLines', 'InternetService',
       'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
       'StreamingTV','StreamingMovies']

# Number of columns in subplot grid
n_cols = 3
n_rows = (len(columns)+n_cols)//n_cols

fig,axes= plt.subplots(n_rows,n_cols,figsize=(15,n_rows*4))

axes=axes.flatten()

for i , col in enumerate(columns):
    sns.countplot(x=col,data= df, ax = axes[i],hue="Churn")
    axes[i].set_title(f'Countplot for {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')

for j in range(i+1,len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

The majority of people who do not churn use services like phoneservicecs , internet services(dsl in sepcific) amd oniline security enabled. For other services like Techsupport , device protection , online backup churn rate is higher when the service is not in use or unavailabe.
